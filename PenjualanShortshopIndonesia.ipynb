```python
plt.figure(figsize=(24,8))
plt.plot(iris['Glove_Nike'],color='r',linestyle='--',linewidth='1.5',label='Glove Nike')
plt.plot(iris['Glove_Rawlings'],color='b',linestyle='--',linewidth='1.5',label='Glove Rawlings')
plt.plot(iris['Glove_Mizuno'],color='m',linestyle='--',linewidth='1.5',label='Glove Mizuno')
plt.plot(iris['Glove_Adidas'],color='g',linestyle='--',linewidth='1.5',label='Glove Adidas')
plt.grid(color='y', linestyle='--',linewidth='1')
plt.style.use('grayscale')
plt.title('Data penjualan Glove bulan Agustus',fontsize=30)
plt.xlabel('Tanggal')
plt.ylabel('banyaknya barang')
plt.legend(loc=1,fontsize=15,labelspacing=0.5)
```
